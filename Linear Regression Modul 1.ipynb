```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/MyDrive/insurance.csv')
```

```python
X = df['bmi'].values
y = df['charges'].values
```

```python
w = 0.0
b = 0.0
learning_rate = 0.0001
epochs = 100
n = len(X)
```

```python
for epoch in range(epochs):
    for i in range(n):
        x_i = X[i]
        y_i = y[i]

        # Prediksi
        y_pred = w * x_i + b

        # Error
        error = y_pred - y_i

        # Gradien
        dw = 2 * error * x_i
        db = 2 * error

        # Update
        w = w - learning_rate * dw
        b = b - learning_rate * db
```

```python
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, w*X + b, color='red', label='Regression Line')
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.legend()
plt.show()
```

```python
# Contoh
x = 30
y = 16000

# Inisialisasi parameter
w = 100
b = 100
learning_rate = 0.001

# Hitung prediksi
y_pred = w * x + b

# Hitung error
epsilon = y_pred - y

# Update parameter
w_new = w - learning_rate * epsilon * x
b_new = b - learning_rate * epsilon

print("===== HASIL SIMULASI SGD =====")
print("Prediksi f(x)      :", y_pred)
print("Error (epsilon)    :", epsilon)
print("w baru             :", w_new)
print("b baru             :", b_new)
```
